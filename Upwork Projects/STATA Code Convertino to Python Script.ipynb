```python
# Stata Code Line 173
df_AMBA_fix = pd.read_excel("Coursematch credit preview report -all programs (AMBA fix).xlsx")
```

```python
df_AMBA_fix.head()
```

```python
# 174 gen existing_credits=total-term
df_AMBA_fix = df_AMBA_fix.assign(existing_credits = df_AMBA_fix['total_enrolled_credits'] - df_AMBA_fix['term_enrolled_credits'])
```

```python
#174  if there is more than one observation per student, just keep one
df_AMBA_fix.drop_duplicates(subset="last_name",
                     keep=False, inplace=True)
```

```python
#178 drop if uni==""
df_AMBA_fix = df_AMBA_fix[df_AMBA_fix.uni != ""]
```

```python
#179-188 add existing credits to rest of the data and save temp file as "scrap"

scrap_AMBA_fix = df_AMBA_fix [['run','last_name','first_name','uni', 
                            'PID', 'program', 'student_term','total_enrolled_credits','term_enrolled_credits',
                            'existing_credits','Requested_Credits','round_type','min_credits','target_credits',
                            'date_inserted','entry_term','processed_by']]

scrap_AMBA_fix.to_excel("scrap_AMBA_fix.xlsx", index=False)
```

```python
#189-204

shortname = pd.DataFrame([df_AMBA_fix.PID,df_AMBA_fix.first_name,df_AMBA_fix.total_enrolled_credits,
                         df_AMBA_fix.Requested_Credits]).transpose()
shortname.head()
```

```python

shortname.rename(columns= {'PID':'student_id','first_name':'shortname',
                                                    'total_enrolled_credits':'allocated', 'Requested_Credits':'pre_enrolled'},inplace = True)
```

```python
shortname = shortname.sort_values(by ='student_id' , ascending=True)
```

```python
shortname.head()
```

```python
shortname.to_excel("students_with_emba_and_mba_versions.xlsx", index=False)
```
