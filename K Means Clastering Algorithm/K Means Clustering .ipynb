```python
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
%matplotlib inline 
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/ML/13_kmeans/income.csv')
data.head()
```

```python
plt.scatter(data['Age'],data['Income($)'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(data[['Age','Income($)']])
y_predicted
```

```python
data['cluster'] = y_predicted
data.head()
```

```python
df1 = data[data.cluster==0]
df2 = data[data.cluster==1]
df3 = data[data.cluster==2]

plt.scatter(df1.Age, df1['Income($)'], color = 'green')
plt.scatter(df2.Age, df2['Income($)'], color = 'red')
plt.scatter(df3.Age, df3['Income($)'], color = 'black')

plt.xlabel('Age')
plt.ylabel('Income($)')
plt.show()
```

```python
scaler = MinMaxScaler()
scaler.fit(data[['Income($)']])
data['Income($)'] = scaler.transform(data[['Income($)']])
scaler.fit(data[['Age']])
data['Age'] = scaler.transform(data[['Age']])
data.head()
```

```python

data
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(data[['Income($)','Age']])
y_predicted
```

```python
data['Cluster'] = y_predicted
data.drop('cluster', axis='columns', inplace=True)
data
```

```python
df1 = data[data.Cluster==0]
df2 = data[data.Cluster==1]
df3 = data[data.Cluster==2]
df4 = data[data.Cluster==3]
df5 = data[data.Cluster==4]


plt.scatter(df1.Age, df1['Income($)'], color = 'green' )
plt.scatter(df2.Age, df2['Income($)'], color = 'red')
plt.scatter(df3.Age, df3['Income($)'], color = 'black')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.show()
```
